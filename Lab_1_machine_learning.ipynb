```python
import random
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
arr_1 = [random.randint(1, 100) for _ in range(50)]
arr_2 = [random.randint(1, 100) for _ in range(50)]
print(arr_1)
print(arr_2)
```

```python
arr = [[0,0] for _ in range(len(arr_1))]

for i in range(len(arr_1)):
  arr[i][0] = arr_1[i]
  arr[i][1] = arr_2[i]
print(arr)
```

```python
plt.scatter(arr_1, arr_2)
plt.show()
```

```python
kmeans = KMeans(3)
kmeans.fit(arr)
```

```python
identified_clusters = kmeans.fit_predict(arr)
identified_clusters
```

```python
plt.scatter(arr_1, arr_2 ,c=identified_clusters,cmap='rainbow')
```

```python
arr_first_colour_x = []
arr_first_colour_y = []

arr_second_colour_x = []
arr_second_colour_y = []

arr_third_colour_x = []
arr_third_colour_y = []

for i in range(len(identified_clusters)):
  if (identified_clusters[i] == 0):
    arr_first_colour_x.append(arr_1[i])
    arr_first_colour_y.append(arr_2[i])
  if (identified_clusters[i] == 1):
    arr_second_colour_x.append(arr_1[i])
    arr_second_colour_y.append(arr_2[i])
  if (identified_clusters[i] == 2):
    arr_third_colour_x.append(arr_1[i])
    arr_third_colour_y.append(arr_2[i])

print(arr_first_colour_x)
print(arr_first_colour_y)
print("-" * 80 )
print(arr_second_colour_x)
print(arr_second_colour_y)
print("-" * 80 )
print(arr_third_colour_x)
print(arr_third_colour_y)
```

```python
arr_first_colour_x_sum = sum(arr_first_colour_x)
arr_first_colour_y_sum = sum(arr_first_colour_y)

arr_second_colour_x_sum = sum(arr_second_colour_x)
arr_second_colour_y_sum = sum(arr_second_colour_y)

arr_third_colour_x_sum = sum(arr_third_colour_x)
arr_third_colour_y_sum = sum(arr_third_colour_y)



arr_first_colour_x_average = arr_first_colour_x_sum / len(arr_first_colour_x)
arr_first_colour_y_average = arr_first_colour_y_sum / len(arr_first_colour_y)

arr_second_colour_x_average = arr_second_colour_x_sum / len(arr_second_colour_x)
arr_second_colour_y_average = arr_second_colour_y_sum / len(arr_second_colour_y)

arr_third_colour_x_average = arr_third_colour_x_sum / len(arr_third_colour_x)
arr_third_colour_y_average = arr_third_colour_y_sum / len(arr_third_colour_y)



print(arr_first_colour_x_average)
print(arr_first_colour_y_average)
print("-" * 30 )
print(arr_second_colour_x_average)
print(arr_second_colour_y_average)
print("-" * 30 )
print(arr_third_colour_x_average)
print(arr_third_colour_y_average)
```

```python
plt.scatter(arr_1, arr_2 ,c=identified_clusters,cmap='rainbow')
plt.scatter(arr_first_colour_x_average, arr_first_colour_y_average)
plt.scatter(arr_second_colour_x_average, arr_second_colour_y_average)
plt.scatter(arr_third_colour_x_average, arr_third_colour_y_average)
plt.show()
```
