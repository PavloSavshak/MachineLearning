```python
import random
import numpy as np
import time
import unittest
```

```python
matrix1 = [[random.randint(0, 1) for _ in range(100)] for _ in range(100)]
matrix2 = [[random.randint(0, 1) for _ in range(100)] for _ in range(100)]
vector1 = [random.randint(0, 1) for _ in range(100)]
vector2 = [random.randint(0, 1) for _ in range(100)]
```

```python
def Numpy(x, y):
    return np.dot(x, y).tolist()
```

```python
def matrix_x_matrix(matrix1, matrix2):
    result = [[0 for _ in range(len(matrix2[0]))] for _ in range(len(matrix1))]
    for i in range(len(matrix1)):
        for j in range(len(matrix2[0])):
            for k in range(len(matrix2)):
                result[i][j] += matrix1[i][k] * matrix2[k][j]
    return result
```

```python
startTime = time.time()
matrix_x_matrix(matrix1, matrix2)
endTime = time.time()
resTime = endTime-startTime
print(f'Час виконання множення матриць без використання Numpy: {resTime}')

startTime = time.time()
Numpy(matrix1, matrix2)
endTime = time.time()
resTime = endTime-startTime
print(f'Час виконання множення матриць з використання Numpy: {resTime}')
```

```python
def matrix_x_vector(matrix, vector):
    result = [0 for _ in range(len(matrix))]
    for i in range(len(matrix)):
        for j in range(len(matrix[0])):
            result[i] += matrix[i][j] * vector[j]
    return result
```

```python
startTime = time.time()
matrix_x_vector(matrix1, vector1)
endTime = time.time()
resTime = endTime-startTime
print(f'Час виконання множення матриці на вектор без використання Numpy: {resTime}')

startTime = time.time()
Numpy(matrix1, vector1)
endTime = time.time()
resTime = endTime-startTime
print(f'Час виконання множення матриці на вектор з використання Numpy: {resTime}')
```

```python
def vector_x_matrix(vector, matrix):
    result = [0 for i in range(len(vector))]
    for i in range(len(matrix)):
        for j in range(len(matrix[0])):
            result[i] += matrix[j][i] * vector[j]
    return result
```

```python
startTime = time.time()
vector_x_matrix(vector1, matrix1)
endTime = time.time()
resTime = endTime-startTime
print(f'Час виконання множення вектора на матрицю без використання Numpy: {resTime}')

startTime = time.time()
Numpy(vector1, matrix1)
endTime = time.time()
resTime = endTime-startTime
print(f'Час виконання множення вектора на матрицю з використання Numpy: {resTime}')
```

```python
def vector_x_vector(vector1, vector2):
    result = 0
    for i in range(len(vector1)):
        result += vector1[i]*vector2[i]
    return result
```

```python
startTime = time.time()
vector_x_vector(vector1, vector2)
endTime = time.time()
resTime = endTime-startTime
print(f'Час виконання множення векторів без використання Numpy: {resTime}')

startTime = time.time()
Numpy(vector1, vector2)
endTime = time.time()
resTime = endTime-startTime
print(f'Час виконання множення векторів з використання Numpy: {resTime}')
```

```python
class Test(unittest.TestCase):
    def test_matrix_x_matrix(self):
        matrix1 = [[random.randint(0, 1) for _ in range(100)] for _ in range(100)]
        matrix2 = [[random.randint(0, 1) for _ in range(100)] for _ in range(100)]
        self.assertEqual(matrix_x_matrix(matrix1, matrix2), Numpy(matrix1, matrix2))

    def test_matrix_x_vector(self):
        matrix = [[random.randint(0, 1) for _ in range(100)] for _ in range(100)]
        vector = [random.randint(0, 1) for _ in range(100)]
        self.assertEqual(matrix_x_vector(matrix, vector), Numpy(matrix, vector))

    def test_vector_x_matrix(self):
        vector = [random.randint(0, 1) for _ in range(100)]
        matrix = [[random.randint(0, 1) for _ in range(100)] for _ in range(100)]
        self.assertEqual(vector_x_matrix(vector, matrix), Numpy(vector, matrix))

    def test_vector_x_vector(self):
        vector1 = [random.randint(0, 1) for _ in range(100)]
        vector2 = [random.randint(0, 1) for _ in range(100)]
        self.assertEqual(vector_x_vector(vector1, vector2), Numpy(vector1, vector2))


unittest.main(argv=[''], verbosity=2, exit=False)
```
